**Visualizing Sine and Cosine Function**

*importing necessary libraries*

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

*assigning X values*

```
X = np.linspace(1,20,100)
```

*plotting sine and cosine curve*

```
plt.plot(X,np.sin(X))
plt.plot(X,np.cos(X))
plt.show()
```
